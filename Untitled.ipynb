```python
import matplotlib.pyplot as plt
```

### 데이터 만들기

```python
a=[1,2,3,4,-5,5,10,-7]
b=['A','B','C','D','E','G','H','I']
```

### 차트로 만들기

```python
plt.bar(b,a,color='red')
```

```python
plt.barh(b,a,color='green')
```

```python
plt.plot(b,a,'r--')
```

```python
plt.plot(b,a,'ro-')
```

```python
plt.plot(b,a,'rx-')
plt.title('subject')
plt.xlabel('school')
plt.ylabel('score')
plt.legend(['jumsu'])
```

### 한글 나타내기

```python
from matplotlib import rc
rc('font',family='batang')
```

```python
plt.plot(b,a,'ro-')
plt.title('과목')
plt.xlabel('학교')
plt.ylabel('점수')
plt.legend(['점수'])
```
